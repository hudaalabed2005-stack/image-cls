```python
import os, json, cv2, numpy as np, pandas as pd
from tqdm import tqdm
from pathlib import Path
from collections import defaultdict

CACHE_DIR = Path("emb_cache"); CACHE_DIR.mkdir(exist_ok=True)

df = pd.read_csv("dataset_index_with_split.csv")
train_df = df[df["split"]=="train"].copy()
val_df   = df[df["split"]=="val"].copy()

class FaceEmbedder:
    def __init__(self):
        import insightface
        from insightface.app import FaceAnalysis
        # Use proper provider name for ORT CPU
        self.fa = FaceAnalysis(name='buffalo_l', providers=['CPUExecutionProvider'])
        self.fa.prepare(ctx_id=-1, det_size=(640,640))
    def embed_path(self, p):
        k = CACHE_DIR / (Path(p).name + ".npy")
        if k.exists(): return np.load(k)
        img = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
        faces = self.fa.get(img)
        if not faces: return None
        f = max(faces, key=lambda z:(z.bbox[2]-z.bbox[0])*(z.bbox[3]-z.bbox[1]))
        e = f.normed_embedding.astype('float32')
        np.save(k, e)
        return e

emb = FaceEmbedder()

MAX_GALLERY_PER_ID = 8   # adjust lower for more speed
MAX_VAL_PER_ID = 4

# Build gallery
gallery = {}
for ident, g in tqdm(train_df.groupby("name"), desc="Gallery"):
    vecs = []
    for p in g["image_path"].tolist()[:MAX_GALLERY_PER_ID]:
        e = emb.embed_path(p)
        if e is not None: vecs.append(e)
    if vecs:
        gallery[ident] = np.stack(vecs).mean(0).astype('float32')

labels = sorted(gallery.keys())
if not labels:
    raise RuntimeError("No gallery identities built.")
G = np.stack([gallery[k] for k in labels]).astype('float32')
os.makedirs("models", exist_ok=True)
np.save("models/gallery_mean.npy", G)
json.dump(labels, open("models/labels.json","w"), indent=2)
print("Gallery identities:", len(labels))

# Build val dict
by = defaultdict(list)
for _, r in val_df.iterrows():
    if r["name"] in gallery:
        by[r["name"]].append(r["image_path"])
for k in list(by.keys()):
    by[k] = by[k][:MAX_VAL_PER_ID]

# Threshold using gallery-vs-val (robust even with 1 val image)
def collect_scores_gallery_vs_val(gallery, by):
    pos, neg = [], []
    names = [n for n in by.keys() if n in gallery and len(by[n]) >= 1]
    if len(names) < 2:
        return np.array([], 'float32'), np.array([], 'float32')
    rng = np.random.default_rng(42)
    name_arr = np.array(names, dtype=object)
    for n in names:
        gvec = gallery[n]
        others = name_arr[name_arr != n]
        if len(others) == 0: continue
        other_g = gallery[rng.choice(others)]
        for p in by[n]:
            e = emb.embed_path(p)
            if e is None: continue
            pos.append(float(np.dot(gvec, e)))
            neg.append(float(np.dot(other_g, e)))
    return np.array(pos, 'float32'), np.array(neg, 'float32')

pos, neg = collect_scores_gallery_vs_val(gallery, by)
print("Positives:", len(pos), "Negatives:", len(neg))

if len(pos)==0 or len(neg)==0:
    thr = 0.35
    print("Not enough scores; using default threshold =", thr)
else:
    scores = np.concatenate([pos, neg])
    labs = np.concatenate([np.ones_like(pos, int), np.zeros_like(neg, int)])
    def find_eer_threshold(scores, labs):
        ths = np.linspace(scores.min(), scores.max(), 200)
        best_th, best_gap = float(scores.mean()), 1e9
        for th in ths:
            pred = (scores >= th).astype(int)
            FAR = ((pred==1)&(labs==0)).sum() / max((labs==0).sum(),1)
            FRR = ((pred==0)&(labs==1)).sum() / max((labs==1).sum(),1)
            gap = abs(FAR-FRR)
            if gap < best_gap:
                best_th, best_gap = float(th), gap
        return best_th
    thr = float(find_eer_threshold(scores, labs))
    print("Computed threshold:", thr)

json.dump({"cosine_threshold": float(thr)}, open("models/threshold.json","w"), indent=2)
print("Saved models/gallery_mean.npy, labels.json, threshold.json")
```

Facenet Embedding Builder (No Alignment)
This code loads the Hugging Face model "lansinuote/simple_facenet" to extract L2-normalized face embeddings directly from RGB images without any face detection or alignment.
It also builds a gallery from a CSV by grouping images per name and optionally aggregating their embeddings (none/mean/median) into a single template per identity.

```python
from pathlib import Path
from typing import Dict, List, Literal, Tuple
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm
import torch
from transformers import AutoModel

Aggregate = Literal["none", "mean", "median"]

class SimpleFacenetNoAlign:

    def __init__(self, model_name: str = "lansinuote/simple_facenet", device: str | None = None, l2_normalize: bool = True):
      # Select device automatically (CUDA if available, otherwise CPU)
        self.device = device or ("cuda" if torch.cuda.is_available() else "cpu")
        # Load HF model in eval mode with float32
        self.model = AutoModel.from_pretrained(model_name, torch_dtype=torch.float32).to(self.device).eval()
        # Whether to L2-normalize output embeddings
        self.l2_normalize = l2_normalize

    @torch.inference_mode()
    def embed_pil(self, img: Image.Image) -> np.ndarray:
       # Convert PIL to RGB tensor, scale to [0,1], then to [-1,1] expected by many face backbones
        x = torch.from_numpy(np.array(img.convert("RGB"))).float()
        x = x.permute(2, 0, 1) / 255.0
        x = (x - 0.5) * 2.0
        x = x.unsqueeze(0).to(self.device)

        out = self.model(x)
       # Try to locate a sensible embedding tensor from common HF outputs
        if hasattr(out, "pooler_output"):
            emb = out.pooler_output
        elif isinstance(out, (list, tuple)):
            emb = out[0]
        elif hasattr(out, "last_hidden_state"):
            v = out.last_hidden_state
            # If spatial dims exist, apply global average pooling; else mean over sequence dim
            emb = v.mean(dim=(2,3)) if v.ndim == 4 else v.mean(dim=1)
        else:
            emb = out
        emb = emb.squeeze(0).detach().cpu().numpy()
        if self.l2_normalize:
            emb = emb / (np.linalg.norm(emb) + 1e-12)
        return emb

def build_gallery_from_metadata(
    metadata_csv: str | Path,
    name_col: str = "name_norm",
    path_col: str = "image_path",
    status_col: str | None = "status",
    valid_status: str = "valid",
    aggregate: Aggregate = "none",
    model_name: str = "lansinuote/simple_facenet",
    l2_normalize: bool = True
) -> Tuple[Dict[str, List[np.ndarray]], Dict[str, np.ndarray], Dict[str, List[str]]]:
    """
    Returns:
      - name_to_embeddings: name -> list of embeddings (only if aggregate='none')
      - name_to_template:  name -> single template (mean/median) or empty if aggregate='none'
      - name_to_paths:     name -> list of image paths actually used
    """
    df = pd.read_csv(metadata_csv)


    if status_col is not None and status_col in df.columns:
        df = df[df[status_col] == valid_status].copy()

    # Optional filtering by status if column exists
    df = df[df[path_col].notna()]
    df = df[df[name_col].notna()]

    extractor = SimpleFacenetNoAlign(model_name=model_name, l2_normalize=l2_normalize)

    name_to_embeddings: Dict[str, List[np.ndarray]] = {}
    name_to_paths: Dict[str, List[str]] = {}

    # Drop rows with missing paths or names
    groups = df.groupby(name_col)[path_col].apply(list).to_dict()

    for name, paths in tqdm(groups.items(), desc="Building gallery"):
        embs: List[np.ndarray] = []
        used_paths: List[str] = []
        for p in paths:
            try:
                img = Image.open(p).convert("RGB")
            except Exception:
                continue
            try:
                emb = extractor.embed_pil(img)
            except Exception:
                continue
            if emb is None or not np.isfinite(emb).all():
                continue
            embs.append(emb)
            used_paths.append(p)

        if len(embs) == 0:
            continue

        name_to_embeddings[name] = embs
        name_to_paths[name] = used_paths

   # Optionally aggregate multiple embeddings into a single template per identity
    name_to_template: Dict[str, np.ndarray] = {}
    if aggregate in ("mean", "median"):
        for name, embs in name_to_embeddings.items():
            arr = np.stack(embs, axis=0)  # N,D
            if aggregate == "mean":
                tpl = arr.mean(axis=0)
            else:
                tpl = np.median(arr, axis=0)
            # تطبيع L2 بعد التجميع
            tpl = tpl / (np.linalg.norm(tpl) + 1e-12)
            name_to_template[name] = tpl

    return name_to_embeddings if aggregate == "none" else {}, name_to_template, name_to_paths

```

 Name Gate for Metadata (Approval Check)
This code loads a CSV of image metadata and provides a “user input gate” to validate a given name against the table.
It returns accepted or rejected based on the status of matching rows (approved/denied) and includes the corresponding image paths.

```python
from pathlib import Path
from typing import Dict, Any, List, Optional
import pandas as pd

def normalize_name_basic(name: str) -> str:
    # Basic normalization: trim extra spaces and lowercase
    return " ".join(name.strip().split()).lower()

def load_metadata(metadata_csv: str | Path) -> pd.DataFrame:
    df = pd.read_csv(metadata_csv)
    return df

def user_input_gate(
    df: pd.DataFrame,
    name_input: str,
    name_col: str = "name",
    status_col: str = "status",
    allowed_status: str = "approved",
    denied_status: str = "denied",
    apply_normalization: bool = False
) -> Dict[str, Any]:

   # Prepare the query name (optional normalization)
    query_name = normalize_name_basic(name_input) if apply_normalization else name_input

    # Filter rows matching the given name
    if apply_normalization:
       # Column values must be normalized similarly
        candidates = df[df[name_col].str.lower() == query_name]
    else:
        candidates = df[df[name_col] == query_name]

    if candidates.empty:
        return {
            "status": "rejected",
            "reason": "Name Not Found",
            "name_input": name_input,
            "approved_image_paths": [],
            "denied_image_paths": []
        }

   # Split candidates by status
    approved_mask = candidates[status_col].astype(str).str.lower() == allowed_status.lower()
    denied_mask   = candidates[status_col].astype(str).str.lower() == denied_status.lower()

    approved_rows = candidates[approved_mask]
    denied_rows   = candidates[denied_mask]

    approved_paths: List[str] = approved_rows["image_path"].dropna().tolist() if "image_path" in candidates.columns else []
    denied_paths:   List[str] = denied_rows["image_path"].dropna().tolist()   if "image_path" in candidates.columns else []

    if len(approved_rows) == 0:
        return {
            "status": "rejected",
            "reason": "All entries are denied for this name",
            "name_input": name_input,
            "approved_image_paths": [],
            "denied_image_paths": denied_paths
        }

    # At least one approved image exists for this name
    return {
        "status": "accepted",
        "reason": "Name has approved images",
        "name_input": name_input,
        "approved_image_paths": approved_paths,
        "denied_image_paths": denied_paths
    }

```

 Simple Facenet Image Embedder
Loads the Hugging Face model "lansinuote/simple_facenet" to convert an input RGB image into a face embedding vector, without any face detection or alignment.
Preprocesses the image to [-1, 1], infers on GPU/CPU, extracts a robust embedding from the model’s outputs, and optionally L2-normalizes it for similarity comparisons.

```python
from typing import Optional
import numpy as np
import torch
from PIL import Image
from transformers import AutoModel

class SimpleFacenetEmbedder:
    def __init__(self, model_name: str = "lansinuote/simple_facenet", device: Optional[str] = None, l2_normalize: bool = True):
        self.device = device or ("cuda" if torch.cuda.is_available() else "cpu")
        self.model = AutoModel.from_pretrained(model_name, torch_dtype=torch.float32).to(self.device).eval()
        self.l2_normalize = l2_normalize

    @torch.inference_mode()
    def embed_image(self, img: Image.Image) -> np.ndarray:

        # Convert PIL image to a tensor in shape [1, 3, H, W] scaled to [-1, 1]
        x = torch.from_numpy(np.array(img.convert("RGB"))).float()
        x = x.permute(2, 0, 1) / 255.0
        x = (x - 0.5) * 2.0
        x = x.unsqueeze(0).to(self.device)

        out = self.model(x)

        # Be robust to different model output structures to obtain an embedding
        if hasattr(out, "pooler_output"):
            emb_t = out.pooler_output                     # 1,D
        elif isinstance(out, (list, tuple)):
            emb_t = out[0]
        elif hasattr(out, "last_hidden_state"):
            v = out.last_hidden_state
            emb_t = v.mean(dim=(2,3)) if v.ndim == 4 else v.mean(dim=1)
        else:
            emb_t = out                                   # 1,D

        emb = emb_t.squeeze(0).detach().cpu().numpy()     # (D,)

        if self.l2_normalize:
            emb = emb / (np.linalg.norm(emb) + 1e-12)
        return emb

```

 Retrieve Embeddings by Name (Full or Template)
Given a person’s name, this function returns the associated face embeddings from either a full-embeddings dictionary or a single-template dictionary.
It optionally prefers all embeddings first, then falls back to a single aggregated template if the full set isn’t available.

```python
from typing import Dict, List, Tuple, Optional
import numpy as np

def get_name_embeddings(
    name_input: str,
    name_to_embeddings: Optional[Dict[str, List[np.ndarray]]] = None,
    name_to_template: Optional[Dict[str, np.ndarray]] = None,
    prefer_all: bool = True
) -> Tuple[str, List[np.ndarray]]:
    # Require at least one source: full embeddings or single-template dictionary
    if name_to_embeddings is None and name_to_template is None:
        raise ValueError("Pass at least one of name_to_embeddings or name_to_template")

    canonical_name = None
    embs: List[np.ndarray] = []

   # Prefer the full list of embeddings if available and requested
    if prefer_all and name_to_embeddings is not None:
        if name_input in name_to_embeddings:
            canonical_name = name_input
            embs = list(name_to_embeddings[name_input])

    # Fallback to the single-template dictionary if nothing found or prefer_all=False
    if (canonical_name is None or len(embs) == 0) and (name_to_template is not None):
        if name_input in name_to_template:
            canonical_name = name_input
            embs = [name_to_template[name_input]]

    # You can add name normalization/matching variants here if needed
    return canonical_name or name_input, embs

```

 Verification Decision via Embedding Distances
Computes distances between a query embedding and a gallery (cosine or Euclidean), then ranks matches.
Applies a decision rule (min, mean_top_k, or voting_k) against a threshold to return accepted/denied, score, and best match info.

```python
from typing import List, Literal, Dict, Any, Optional, Tuple
import numpy as np

Similarity = Literal["cosine", "euclidean"]
DecisionRule = Literal["min", "mean_top_k", "voting_k"]

def cosine_distance(a: np.ndarray, b: np.ndarray) -> float:
    # Cosine distance = 1 - cosine similarity (smaller is better
    denom = (np.linalg.norm(a) * np.linalg.norm(b)) + 1e-12
    return 1.0 - float(np.dot(a, b) / denom)

def euclidean_distance(a: np.ndarray, b: np.ndarray) -> float:
    return float(np.linalg.norm(a - b))

def compute_distances(
    query_emb: np.ndarray,
    gallery_embs: List[np.ndarray],
    similarity: Similarity = "cosine"
) -> np.ndarray:
    if len(gallery_embs) == 0:
        return np.array([])
    if similarity == "cosine":
        dists = [cosine_distance(query_emb, g) for g in gallery_embs]
    elif similarity == "euclidean":
        dists = [euclidean_distance(query_emb, g) for g in gallery_embs]
    else:
        raise ValueError("similarity must be 'cosine' or 'euclidean'")
    return np.array(dists)

def decide_verification(
    query_emb: np.ndarray,
    gallery_embs: List[np.ndarray],
    gallery_paths: Optional[List[str]] = None,
    similarity: Similarity = "cosine",
    rule: DecisionRule = "min",
    threshold: float = 0.35,
    k: int = 3
) -> Dict[str, Any]:

    dists = compute_distances(query_emb, gallery_embs, similarity=similarity)
    if dists.size == 0:
        return {"status": "denied", "reason": "No gallery embeddings", "score": None, "best_match_path": None, "details": {}}

    sort_idx = np.argsort(dists)  # Ascending (smallest distance first)
    dists_sorted = dists[sort_idx]

    # Track the path of the closest match if provided
    best_match_path = None
    if gallery_paths is not None and len(gallery_paths) == len(gallery_embs):
        best_match_path = gallery_paths[int(sort_idx[0])]

   # Compute the decision score according to the selected rule
    if rule == "min":
        score = float(dists_sorted[0])  # أصغر مسافة
        accepted = score <= threshold

    elif rule == "mean_top_k":
        kk = min(k, len(dists_sorted))
        score = float(dists_sorted[:kk].mean())  # متوسط أصغر k مسافات
        accepted = score <= threshold

    elif rule == "voting_k":
        kk = min(k, len(dists_sorted))

        within = (dists_sorted[:kk] <= threshold).sum()
        score = float(within) / kk
        accepted = within > (kk // 2)
    else:
        raise ValueError("rule must be 'min', 'mean_top_k', or 'voting_k'")

    result = {
        "status": "accepted" if accepted else "denied",
        "score": score,
        "best_match_path": best_match_path,
        "details": {
            "similarity": similarity,
            "rule": rule,
            "threshold": threshold,
            "all_distances": dists.tolist(),
            "sorted_indices": sort_idx.tolist(),
            "top_k": int(min(k, len(dists_sorted))),
            "min_distance": float(dists_sorted[0])
        }
    }
    return result

```

Builds genuine and impostor distance distributions from embeddings, then computes FAR/FRR across thresholds.
Selects an optimal decision threshold using either Equal Error Rate (EER) or minimal average error, returning detailed metrics.

```python
from typing import Dict, List, Literal, Tuple, Optional
import numpy as np
import itertools

Similarity = Literal["cosine", "euclidean"]

def cosine_distance(a: np.ndarray, b: np.ndarray) -> float:
  # Cosine distance = 1 - cosine similarity (smaller is better)
    denom = (np.linalg.norm(a) * np.linalg.norm(b)) + 1e-12
    return 1.0 - float(np.dot(a, b) / denom)

def euclidean_distance(a: np.ndarray, b: np.ndarray) -> float:
    return float(np.linalg.norm(a - b))

def pairwise_distances(
    name_to_embeddings: Dict[str, List[np.ndarray]],
    similarity: Similarity = "cosine",
    max_pairs_per_class: Optional[int] = None,
    max_impostor_pairs: int = 200000
) -> Tuple[np.ndarray, np.ndarray]:
   # Build distributions of genuine and impostor distances for calibration/evaluation
    rng = np.random.default_rng(42)
    # Pick distance function
    dist_fn = cosine_distance if similarity == "cosine" else euclidean_distance

    # Genuine pairs
    genuine = []
    for name, embs in name_to_embeddings.items():
        if len(embs) < 2:
            continue
        idxs = list(range(len(embs)))
        pairs = list(itertools.combinations(idxs, 2))
        if max_pairs_per_class is not None and len(pairs) > max_pairs_per_class:
            pairs = list(rng.choice(pairs, size=max_pairs_per_class, replace=False))
        for i, j in pairs:
            d = dist_fn(embs[i], embs[j])
            genuine.append(d)
    genuine_dists = np.array(genuine, dtype=np.float32)

    # Impostor pairs: pairs drawn across different identities (random sampling)
    names = [n for n, v in name_to_embeddings.items() if len(v) > 0]
    impostor = []
    if len(names) >= 2:
        # Precompute indices for each name, then sample random cross-identity pairs
        name_to_idx = {n: list(range(len(name_to_embeddings[n]))) for n in names}
        count = 0
        while count < max_impostor_pairs:
            n1, n2 = rng.choice(names, size=2, replace=False)
            i = int(rng.choice(name_to_idx[n1]))
            j = int(rng.choice(name_to_idx[n2]))
            d = dist_fn(name_to_embeddings[n1][i], name_to_embeddings[n2][j])
            impostor.append(d)
            count += 1
    impostor_dists = np.array(impostor, dtype=np.float32)

    return genuine_dists, impostor_dists

def compute_far_frr(
    genuine_dists: np.ndarray,
    impostor_dists: np.ndarray,
    thresholds: np.ndarray
) -> Tuple[np.ndarray, np.ndarray]:
    """
    FAR(th) = fraction of impostor_dists <= th  (false accept rate)
    FRR(th) = fraction of genuine_dists >  th   (false reject rate)
    """
    # Sort once for fast threshold sweeps
    imp_sorted = np.sort(impostor_dists)
    gen_sorted = np.sort(genuine_dists)

    fars = np.searchsorted(imp_sorted, thresholds, side="right") / max(len(imp_sorted), 1)
    frrs = (len(gen_sorted) - np.searchsorted(gen_sorted, thresholds, side="right")) / max(len(gen_sorted), 1)

    return fars, frrs

def find_best_threshold(
    genuine_dists: np.ndarray,
    impostor_dists: np.ndarray,
    strategy: Literal["eer", "min_avg"] = "eer",
    num_thresholds: int = 2001
) -> Tuple[float, dict]:
    """
    - strategy "eer": pick threshold where |FAR - FRR| is minimized (Equal Error Rate point).
    - strategy "min_avg": pick threshold minimizing (FAR + FRR)/2.
    """
    if len(genuine_dists) == 0 or len(impostor_dists) == 0:
        raise ValueError("Need both genuine and impostor distances for calibration.")

    dmin = float(min(genuine_dists.min(), impostor_dists.min()))
    dmax = float(max(genuine_dists.max(), impostor_dists.max()))
    thresholds = np.linspace(dmin, dmax, num_thresholds, dtype=np.float32)

    fars, frrs = compute_far_frr(genuine_dists, impostor_dists, thresholds)

    if strategy == "eer":
        idx = int(np.argmin(np.abs(fars - frrs)))
    else:  # "min_avg"
        idx = int(np.argmin((fars + frrs) / 2.0))

    best_th = float(thresholds[idx])
    info = {
        "threshold": best_th,
        "FAR": float(fars[idx]),
        "FRR": float(frrs[idx]),
        "diff": float(abs(fars[idx] - frrs[idx])),
        "avg": float(((fars[idx] + frrs[idx]) / 2.0)),
        "thresholds": thresholds,
        "FARs": fars,
        "FRRs": frrs,
    }
    return best_th, info

```

End-to-End Face Verification Pipeline (No Alignment)
Loads "lansinuote/simple_facenet" to embed an input image, gates by name/status from metadata, and compares the query embedding to a gallery using cosine or Euclidean distances.
Applies decision rules (min, mean_top_k, voting_k) with a threshold to return accepted/denied, score, and the best-matching image path.

```python
from typing import Dict, List, Any, Optional, Tuple, Literal
from dataclasses import dataclass
import numpy as np
import pandas as pd
from PIL import Image
import torch
from transformers import AutoModel

# ---------- 1) Embedder ----------
class SimpleFacenetEmbedder:
    def __init__(self, model_name: str = "lansinuote/simple_facenet", device: Optional[str] = None, l2_normalize: bool = True):
        self.device = device or ("cuda" if torch.cuda.is_available() else "cpu")
        self.model = AutoModel.from_pretrained(model_name, torch_dtype=torch.float32).to(self.device).eval()
        self.l2_normalize = l2_normalize

    @torch.inference_mode()
    def embed_image(self, img: Image.Image) -> np.ndarray:
        x = torch.from_numpy(np.array(img.convert("RGB"))).float()  # H,W,3
        x = x.permute(2, 0, 1) / 255.0                              # 3,H,W in [0,1]
        x = (x - 0.5) * 2.0                                         # [-1,1]
        x = x.unsqueeze(0).to(self.device)                          # 1,3,H,W

        out = self.model(x)
        if hasattr(out, "pooler_output"):
            emb_t = out.pooler_output
        elif isinstance(out, (list, tuple)):
            emb_t = out[0]
        elif hasattr(out, "last_hidden_state"):
            v = out.last_hidden_state
            emb_t = v.mean(dim=(2,3)) if v.ndim == 4 else v.mean(dim=1)
        else:
            emb_t = out

        emb = emb_t.squeeze(0).detach().cpu().numpy()
        if self.l2_normalize:
            emb = emb / (np.linalg.norm(emb) + 1e-12)
        return emb

# ---------- 2) Name/status gate to quickly filter by metadata ----------
def user_input_gate(
    df: pd.DataFrame,
    name_input: str,
    name_col: str = "name",
    status_col: str = "status",
    allowed_status: str = "approved",
    denied_status: str = "denied",
) -> Dict[str, Any]:
    candidates = df[df[name_col] == name_input]
    if candidates.empty:
        return {"status": "rejected", "reason": "Name Not Found", "approved_paths": [], "denied_paths": []}

    approved_rows = candidates[candidates[status_col].astype(str).str.lower() == allowed_status.lower()]
    denied_rows   = candidates[candidates[status_col].astype(str).str.lower() == denied_status.lower()]

    approved_paths = approved_rows["image_path"].dropna().tolist() if "image_path" in candidates.columns else []
    denied_paths   = denied_rows["image_path"].dropna().tolist()   if "image_path" in candidates.columns else []

    if len(approved_rows) == 0:
        return {"status": "rejected", "reason": "All entries are denied for this name", "approved_paths": [], "denied_paths": denied_paths}

    return {"status": "accepted", "reason": "ok", "approved_paths": approved_paths, "denied_paths": denied_paths}

# ---------- 3) Distance computation and decision logic -----------
Similarity = Literal["cosine", "euclidean"]
DecisionRule = Literal["min", "mean_top_k", "voting_k"]

def cosine_distance(a: np.ndarray, b: np.ndarray) -> float:
    denom = (np.linalg.norm(a) * np.linalg.norm(b)) + 1e-12
    return 1.0 - float(np.dot(a, b) / denom)

def euclidean_distance(a: np.ndarray, b: np.ndarray) -> float:
    return float(np.linalg.norm(a - b))

def compute_distances(query_emb: np.ndarray, gallery_embs: List[np.ndarray], similarity: Similarity = "cosine") -> np.ndarray:
    if similarity == "cosine":
        return np.array([cosine_distance(query_emb, g) for g in gallery_embs], dtype=np.float32)
    elif similarity == "euclidean":
        return np.array([euclidean_distance(query_emb, g) for g in gallery_embs], dtype=np.float32)
    else:
        raise ValueError("similarity must be 'cosine' or 'euclidean'")

def decide_verification(
    query_emb: np.ndarray,
    gallery_embs: List[np.ndarray],
    gallery_paths: Optional[List[str]] = None,
    similarity: Similarity = "cosine",
    rule: DecisionRule = "min",
    threshold: float = 0.35,
    k: int = 3
) -> Dict[str, Any]:
    dists = compute_distances(query_emb, gallery_embs, similarity=similarity)
    if dists.size == 0:
        return {"status": "denied", "reason": "No gallery embeddings", "score": None, "matched_image_path": None}

    sort_idx = np.argsort(dists)
    d_sorted = dists[sort_idx]
    best_idx = int(sort_idx[0])
    best_path = gallery_paths[best_idx] if gallery_paths and len(gallery_paths) == len(gallery_embs) else None

    if rule == "min":
        score = float(d_sorted[0])
        accepted = score <= threshold
    elif rule == "mean_top_k":
        kk = min(k, len(d_sorted))
        score = float(d_sorted[:kk].mean())
        accepted = score <= threshold
    else:  # voting_k
        kk = min(k, len(d_sorted))
        within = int((d_sorted[:kk] <= threshold).sum())
        score = within / kk
        accepted = within > (kk // 2)

    return {
        "status": "accepted" if accepted else "denied",
        "score": score,
        "matched_image_path": best_path,
        "details": {
            "min_distance": float(d_sorted[0]),
            "all_distances": dists.tolist(),
            "similarity": similarity,
            "rule": rule,
            "threshold": threshold
        }
    }

# ---------- 4) End-to-end verify() that returns a payload ----------
def verify_user(
    df_metadata: pd.DataFrame,
    name_input: str,
    input_image_path: str,
    name_to_embeddings: Dict[str, List[np.ndarray]],
    name_to_paths: Dict[str, List[str]],
    embedder: SimpleFacenetEmbedder,
    similarity: Similarity = "cosine",
    rule: DecisionRule = "min",
    threshold: float = 0.35,
    name_col: str = "name",
    status_col: str = "status",
    allowed_status: str = "approved",
    denied_status: str = "denied"
) -> Dict[str, Any]:
    # 1) Quick name/status gate using metadata
    gate = user_input_gate(df_metadata, name_input, name_col, status_col, allowed_status, denied_status)
    if gate["status"] != "accepted":
        return {
            "name": name_input,
            "status": "denied",
            "score": None,
            "matched_image_path": None,
            "reason": gate.get("reason", "Rejected by gate")
        }

    # 2) Read input image and extract query embedding
    try:
        img = Image.open(input_image_path).convert("RGB")
    except Exception as e:
        return {"name": name_input, "status": "denied", "score": None, "matched_image_path": None, "reason": f"Image read error: {e}"}

    query_emb = embedder.embed_image(img)
    if query_emb is None or not np.isfinite(query_emb).all():
        return {"name": name_input, "status": "denied", "score": None, "matched_image_path": None, "reason": "Embedding failed"}

    # 3) Fetch gallery embeddings for this name
    if name_input not in name_to_embeddings or len(name_to_embeddings[name_input]) == 0:
        return {"name": name_input, "status": "denied", "score": None, "matched_image_path": None, "reason": "No gallery for this name"}

    gallery_embs = name_to_embeddings[name_input]
    gallery_paths = name_to_paths.get(name_input, None)

    # 4) Decision stage
    decision = decide_verification(
        query_emb=query_emb,
        gallery_embs=gallery_embs,
        gallery_paths=gallery_paths,
        similarity=similarity,
        rule=rule,
        threshold=threshold,
        k=3
    )

    payload = {
        "name": name_input,
        "status": decision["status"],
        "score": decision["score"],
        "matched_image_path": decision.get("matched_image_path"),
        "reason": None if decision["status"] == "accepted" else "Not within threshold",
    }
    # Attach details for debugging/analysis
    payload["details"] = decision.get("details", {})
    return payload
```

Minimal Gradio Face Check (Text-Only)
Loads a CSV to map each name to existing image paths, then uses a simple handcrafted embedding (color stats + gradients) with cosine distance to compare the input image to that name’s gallery.
A Gradio UI takes a name and image, computes the best match score, and returns Approved/Denied text with the score and an approximate accuracy.

```python
# !pip install -U gradio pillow numpy pandas tqdm opencv-python-headless

import os
import numpy as np
import pandas as pd
from PIL import Image
from typing import Dict, List
import gradio as gr

#General settings
CSV_PATH = "dataset_index.csv"
NAME_COL = "name"
IMAGE_COL = "image_path"
STATUS_COL = "status"
ALLOWED_STATUS = "approved"
BASE_DIR = None
THRESHOLD = 0.35

# 1) Load CSV and build mapping: name -> list of image paths
assert os.path.exists(CSV_PATH), f"CSV not found: {CSV_PATH}"
df = pd.read_csv(CSV_PATH)

if STATUS_COL in df.columns:
    df_use = df[df[STATUS_COL].astype(str).str.lower() == ALLOWED_STATUS.lower()].copy()
else:
    df_use = df.copy()

df_use = df_use[[NAME_COL, IMAGE_COL]].dropna()

def resolve_path(p: str) -> str:
    p = str(p).strip()
    if BASE_DIR and not os.path.isabs(p):
        return os.path.join(BASE_DIR, p)
    return p

df_use[IMAGE_COL] = df_use[IMAGE_COL].apply(resolve_path)

name_to_paths: Dict[str, List[str]] = {}
for _, row in df_use.iterrows():
    name = str(row[NAME_COL]).strip()
    path = str(row[IMAGE_COL]).strip()
    if os.path.exists(path):
        name_to_paths.setdefault(name, []).append(path)

# 2) Very simple features + cosine distance
def simple_embedding(img: Image.Image, size: int = 64) -> np.ndarray:
    arr = img.convert("RGB").resize((size, size), Image.BILINEAR)
    arr = np.asarray(arr, dtype=np.float32)  # H,W,3
    mean_colors = arr.mean(axis=(0,1))       # 3
    std_colors = arr.std(axis=(0,1)) + 1e-6  # 3
    gx = np.abs(np.diff(arr, axis=1)).mean()
    gy = np.abs(np.diff(arr, axis=0)).mean()
    feat = np.concatenate([mean_colors, std_colors, [gx, gy]], axis=0).astype(np.float32)
    feat = feat / (np.linalg.norm(feat) + 1e-12)
    return feat

def cosine_distance(a: np.ndarray, b: np.ndarray) -> float:
    return 1.0 - float(np.dot(a, b) / ((np.linalg.norm(a) * np.linalg.norm(b)) + 1e-12))

# 3) Verification function: returns text only
def verify_minimal_text_only(name_input: str, input_image: Image.Image):
    name_input = (name_input or "").strip()
    if not name_input:
        return "Please enter a name."
    if input_image is None:
        return "Please upload an image."

    gallery_paths = name_to_paths.get(name_input, [])
    if len(gallery_paths) == 0:
        return "Denied: The verification request was rejected. (reason=Name Not Found)"

    q = simple_embedding(input_image)
    best_score = None

    for p in gallery_paths:
        try:
            gimg = Image.open(p).convert("RGB")
        except Exception:
            continue
        g = simple_embedding(gimg)
        d = cosine_distance(q, g)
        if best_score is None or d < best_score:
            best_score = d

    if best_score is None:
        return "Denied: The verification request was rejected. (reason=No readable gallery images)"

    accepted = best_score <= THRESHOLD
    accuracy = max(0.0, 1.0 - float(best_score)) * 100.0

    if accepted:
        return f"Approved: The verification request was accepted. (score={best_score:.3f}, accuracy={accuracy:.1f}%)"
    else:
        return f"Denied: The verification request was rejected. (score={best_score:.3f}, accuracy={accuracy:.1f}%)"

# 4) Gradio UI: name + image -> text-only result
with gr.Blocks() as demo:
    gr.Markdown("Minimal Face Check (text-only output)")
    name_in = gr.Textbox(label=f"Name ({NAME_COL} in CSV)", placeholder=f"اكتب {NAME_COL} كما في CSV")
    img_in = gr.Image(label="Input image", type="pil")
    btn = gr.Button("Verify")
    out_text = gr.Textbox(label="Result", lines=2)
    btn.click(fn=verify_minimal_text_only, inputs=[name_in, img_in], outputs=[out_text])

demo.launch(share=False)
```

```python
# إعدادات أعمدتك (كما في كودك)
NAME_COL = "name"
IMAGE_COL = "image_path"

# مثال تدريب عام (كي تتبنّاه مع موديلك و criterion و optimizer الحاليين)
def run_training(num_epochs, model, criterion, optimizer, device, train_loader, val_loader):
    logger = TrainCurveLogger()

    for epoch in range(1, num_epochs + 1):
        # ===== Train split =====
        model.train()
        train_running_loss = 0.0
        train_correct = 0
        train_total = 0

        for batch in train_loader:
            # يفترض أن اللودِر يعيد صور وتسمية صحيحة
            imgs = batch["image"].to(device) if isinstance(batch, dict) else batch[0].to(device)
            labels = batch["label"].to(device) if isinstance(batch, dict) else batch[1].to(device)

            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            bs = labels.size(0)
            train_running_loss += loss.item() * bs

            preds = outputs.argmax(dim=1)
            train_correct += (preds == labels).sum().item()
            train_total += bs

        train_loss_epoch = train_running_loss / max(train_total, 1)
        train_acc_epoch = train_correct / max(train_total, 1)

        # ===== Val split =====
        model.eval()
        val_running_loss = 0.0
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for batch in val_loader:
                imgs = batch["image"].to(device) if isinstance(batch, dict) else batch[0].to(device)
                labels = batch["label"].to(device) if isinstance(batch, dict) else batch[1].to(device)

                outputs = model(imgs)
                loss = criterion(outputs, labels)

                bs = labels.size(0)
                val_running_loss += loss.item() * bs

                preds = outputs.argmax(dim=1)
                val_correct += (preds == labels).sum().item()
                val_total += bs

        val_loss_epoch = val_running_loss / max(val_total, 1)
        val_acc_epoch = val_correct / max(val_total, 1)

        # تسجيل القيم باللوجر
        logger.log(
            epoch=epoch,
            train_loss=train_loss_epoch,
            val_loss=val_loss_epoch,
            train_acc=train_acc_epoch,
            val_acc=val_acc_epoch
        )

        # طباعة مختصرة باسماء أعمدتك المرجعية
        print(f"[Epoch {epoch}] loss(train/val) = {train_loss_epoch:.4f}/{val_loss_epoch:.4f} | acc(train/val) = {train_acc_epoch:.4f}/{val_acc_epoch:.4f}")

    # رسم المنحنيات
    logger.plot(show_acc=True)
    return logger
```

```python
# !pip install -U torch torchvision pillow matplotlib numpy

import torch
import torch.nn.functional as F
from torchvision import transforms
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model.eval()
        self.target_layer = target_layer
        self.activations = None
        self.gradients = None

        # forward hook to cache activations
        self.fwd_handle = target_layer.register_forward_hook(self._save_activation)
        # backward hook to cache gradients
        self.bwd_handle = target_layer.register_full_backward_hook(self._save_gradient)

    def _save_activation(self, module, inp, out):
        self.activations = out.detach()

    def _save_gradient(self, module, grad_in, grad_out):
        # grad_out is a tuple; take grad w.r.t. output of target layer
        self.gradients = grad_out[0].detach()

    def __del__(self):
        # clean up hooks
        try:
            self.fwd_handle.remove()
            self.bwd_handle.remove()
        except Exception:
            pass

    @torch.inference_mode(False)
    def generate(self, input_tensor, class_idx=None):
        """
        input_tensor: shape [1, 3, H, W]
        class_idx: target class index for which to compute Grad-CAM.
                   If None, uses argmax of model logits.
        Returns: heatmap (H, W) in [0, 1]
        """
        self.model.zero_grad(set_to_none=True)
        logits = self.model(input_tensor)  # [1, C]
        if class_idx is None:
            class_idx = int(logits.argmax(dim=1).item())

        score = logits[0, class_idx]
        score.backward(retain_graph=True)

        # activations: [B, C, H', W'], gradients: [B, C, H', W']
        act = self.activations
        grad = self.gradients
        weights = grad.mean(dim=(2, 3), keepdim=True)  # [B, C, 1, 1]
        cam = (weights * act).sum(dim=1, keepdim=False)  # [B, H', W']
        cam = F.relu(cam)
        # normalize to [0,1]
        cam = cam - cam.min()
        cam = cam / (cam.max() + 1e-12)

        # upsample to input size
        cam_up = F.interpolate(cam.unsqueeze(1), size=input_tensor.shape[-2:], mode="bilinear", align_corners=False)
        return cam_up[0, 0].detach().cpu().numpy(), class_idx

def overlay_heatmap(img_pil, heatmap, alpha=0.35, cmap="jet"):
    """
    img_pil: original RGB PIL.Image
    heatmap: numpy array (H, W) in [0,1]
    Returns: blended RGB image as numpy array
    """
    img = np.asarray(img_pil.convert("RGB")).astype(np.float32) / 255.0
    hmap = plt.get_cmap(cmap)(heatmap)[..., :3]  # RGB
    blended = (1 - alpha) * img + alpha * hmap
    blended = np.clip(blended, 0, 1)
    return blended

# Example usage (adapt to your model and preprocessing)
# 1) Define your model and choose a target conv layer
# Example with torchvision resnet18
# from torchvision.models import resnet18
# model = resnet18(weights=None)  # or with weights
# target_layer = model.layer4[1].conv2  # a deep conv layer

# 2) Preprocess matching your training pipeline
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),                     # [0,1]
    transforms.Normalize(mean=[0.485,0.456,0.406],
                         std=[0.229,0.224,0.225])
])

def run_gradcam_on_image(model, target_layer, image_path, class_idx=None, class_names=None, device=None):
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device).eval()
    img_pil = Image.open(image_path).convert("RGB")
    inp = preprocess(img_pil).unsqueeze(0).to(device)

    cam_engine = GradCAM(model, target_layer)
    heatmap, used_class = cam_engine.generate(inp, class_idx=class_idx)

    blended = overlay_heatmap(img_pil.resize(heatmap.shape[::-1]), heatmap, alpha=0.35)
    title = f"Grad-CAM for class {used_class}"
    if class_names and 0 <= used_class < len(class_names):
        title = f"Grad-CAM: {class_names[used_class]} (id={used_class})"

    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1); plt.imshow(img_pil); plt.axis("off"); plt.title("Original")
    plt.subplot(1,2,2); plt.imshow(blended); plt.axis("off"); plt.title(title)
    plt.tight_layout(); plt.show()

```
